```python
def prepare_text(text):
    sent = ' '.join(map(lambda x: "\<p>"+x+"\</p>", 
                        map(str.strip, 
                            filter(len, 
                                   ' '.join(map(lambda string: re.sub(r"\xa0", " ", re.sub('\\\\"','"', string)), 
                                                filter( lambda x: "Section::::" not in x, 
                                                       map(str.strip, 
                                                           map(lambda x:x+'.', 
                                                               filter(len, 
                                                                      BeautifulSoup(html.unescape(text.strip('\" '))).get_text().split('.'))
                                                              )
                                                          )
                                                      )
                                               )
                                           ).split('\\n')
                                  ) 
                           )
                       )
                   )
    return sent
```
